## Data Visualization represents relationship btw data through graphical, charts, graphs view using two main libraries


- [Matplotlib](https://matplotlib.org/): Plotting and visualization library for Python. We'll use the `pyplot` module from `matplotlib`. As convention, it is often imported as `plt`.
- [Seaborn](https://seaborn.pydata.org/): An easy-to-use visualization library that builds on top of Matplotlib and lets you create beautiful charts with just a few lines of code.


```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

### we can style the graphs using [rcParam](https://matplotlib.org/3.2.1/tutorials/introductory/customizing.html)

```python
sns.set_style("darkgrid")
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['figure.facecolor'] = '#00000000'
```

## Line Chart
Used to represent data points(X vs Y axis) and connecting them through line
We can customize lines, shape, colour, font etc

```python
#consider line chart of watermelon and oranges rates depending on KG where more the weight there will be some discount
#x-axis
weights=range(1,11)

#y-axis
price_watermelon=[120,200,280,340,360,400,435,470,500,515]
price_oranges=[12,24,35,45,56,70,80,92,102,110]

#Line1
plt.plot(weights,price_watermelon,'g-x',linewidth=2,markersize=12,markeredgewidth=4,markeredgecolor='black')

#Line2
plt.plot(weights,price_oranges,'r--o',linewidth=2,markersize=6,markeredgewidth=1,markeredgecolor='black')

#title
plt.title('Weights of fruits vs price')

#line labels
plt.legend(['water melon','oranges'])

#axis labels
plt.xlabel('weights(KG)');plt.ylabel('price($)')
```

## Scatter plot
It is used to plot graph btw 2 variables and additional third variable can also be used to determine points belogning to that third variable

```python
data=sns.load_dataset('iris')
data.sample(5)
```

```python
sns.scatterplot(data.sepal_length,#X-axis
                data.sepal_width,#y-axis
                hue=data.species,#3-rd variable
                s=100 )#size of point
plt.title('Flowers')
```

## Histogram and Frequency distribution

Grpah of a variable showing how many (frequency of that individual) falls into that value of variable.                     
eg: 5 flowers(frequency) have(falls into value) sepal_length 20

```python
plt.title('Distribution of sepal length')
sns.distplot(data.sepal_length,kde=False)
```

###  We can also look at this data as a frequency distribution, where the values on Y-axis are percentagess instead of counts.

 

```python
plt.title('Distribution of sepal length')
sns.distplot(data.sepal_length)
```

## HeatMap
Graphical representation of 2-D data using colors
Uses: Confusion matrix

```python
#Flight dataset
flights=sns.load_dataset('flights').pivot('year','month','passengers')

#heatmap
plt.title("no of passenegrs(1000s)")

sns.heatmap(flights, 
            fmt="d", 
            annot=True, 
            linewidths=.5, 
            cmap='Blues',
            annot_kws={"fontsize":14});
```

## Contour Plot
It uses contours or color coded regions helps us to visualize 3 numerical variables in 2 dimensions.<br>
one variable- Horizontal axis<br>
second variable- Vertical axis<br>
third variable- represented by color gradient and isolines(lines of constant values)


```python
plt.title("flowers")
sns.kdeplot(data.sepal_length,data.sepal_width,shade=True,shade_lowest=False)
```

#### We can segment speicies of flowers by creating multiple contour plots with different colors. 

```python
setosa=data[data.species== 'setosa']
virginica=data[data.species== 'virginica']
plt.title("flowers(setosa,virginica)")
sns.kdeplot(setosa.sepal_length,setosa.sepal_width,shade=True,cmap="Reds",shade_lowest=False)
sns.kdeplot(virginica.sepal_length,virginica.sepal_width,shade=True,cmap="Blues",shade_lowest=False);
```

## Box Plot
A box plot shows the distribution of data along a single axis, using a "box" and "whiskers". The lower end of the box represents the 1st quartile (i.e. 25% of values are below it), and the upper end of the box represents the 3rd quartile (i.e. 25% of values are above it). The median value is represented via a line inside the box. The "whiskers" represent the minimum & maximum values (sometimes excluding outliers, which are represented as dots).

### Example
We'll use another sample dataset included with Seaborn, called "tips". The dataset contains information about the sex, time of day, total bill and tip amount for customers visiting a restraurant over a week.

```python
tips=sns.load_dataset('tips')
tips.sample(5)
```

```python
plt.title("daily total bill")

sns.boxplot(tips.day,tips.total_bill,hue=tips.smoker)
```

## Bar Chart
A bar chart presents categorical data with rectangular bars with heights proportional to the values that they represent. If there are multiple values for each category, then a bar plot can also represent the average value, with confidence intervals.

### Example
We can use a bar chart visulize the average value of total bill for different days of the week, segmented by sex, for the "tips" dataset 

```python
sns.barplot(x="day",y="total_bill",hue="sex",data=tips)
```
